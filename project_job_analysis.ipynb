```python
inner_text=[]
href_link_list=[]
soup_list=[]

for i in range(len(url_list)):
    browser.get(url_list[i])
    browser.implicitly_wait(3)
    
    webpage=requests.get(url_list[i]).text
    
    soup_list.append(BeautifulSoup(webpage,'lxml'))
    
    for link in soup_list[i].find_all("a"):
        href_link_list.append(link.get("href"))
        inner_text.append("Inner Text: {}".format(link.text.strip()))
```

```python
href_link_final=[]
for i in range(len(href_link_list)):
    if 'jobs/view' in str(href_link_list[i]):
        href_link_final.append(href_link_list[i])
    else:
        continue
        
```

```python
len(href_link_list),len(href_link_final)
```

```python
Job_title=[]
Company_name=[]
Location_name=[]
Time_posted=[]
Applicants_count=[]
Job_nature_level=[]
Emp_count_Ctype=[]
Emp_count=[]
Emp_on_Linkedin=[]
Pay_range=[]
Company_followers=[]
```

```python
count=0
for link in range(90,len(href_link_final)):
    count+=1
    print(count)
    browser.get(href_link_final[link])
    browser.implicitly_wait(5)
    
    try:
        job_title=browser.find_elements_by_class_name("jobs-unified-top-card__job-title")
        Job_title.append(job_title[0].text)
    except:
        Job_title.append('NA')
    
    try:
        company_name=browser.find_elements_by_class_name("jobs-unified-top-card__company-name")
        Company_name.append(company_name[0].text)
    except:
        Company_name.append('NA')
        
    try:
        location_name=browser.find_elements_by_class_name("jobs-unified-top-card__bullet")
        Location_name.append(location_name[0].text)
    except:
        Location_name.append('NA')
    
    try:
        time_posted=browser.find_elements_by_class_name("jobs-unified-top-card__posted-date")
        Time_posted.append(time_posted[0].text)
    except:
        Time_posted.append('NA')
        
    try:
        applicants_count=browser.find_elements_by_class_name("jobs-unified-top-card__applicant-count")
        Applicants_count.append(applicants_count[0].text)
    except:
        Applicants_count.append('NULL')
        
    try:
        job_nature_level=browser.find_elements_by_class_name("jobs-unified-top-card__job-insight")
        Job_nature_level.append(job_nature_level[0].text)
        Emp_count_Ctype.append(job_nature_level[1].text)
    except:
        Job_nature_level.append('NA')
        Emp_count_Ctype.append('NA')
        
    try:
        emp_count_onLinkedin=browser.find_elements_by_class_name("jobs-company__inline-information")
        Emp_count.append(emp_count_onLinkedin[0].text)
        Emp_on_Linkedin.append(emp_count_onLinkedin[1].text)
        
    except:
        Emp_count.append('NA')
        Emp_on_Linkedin.append('NA')
        
    try:
        pay_range=browser.find_elements_by_class_name("jobs-unified-description__salary-main-rail-card")
        Pay_range.append(pay_range[0].text)
    except:
        Pay_range.append('NA')
        
    try:
        followers=browser.find_elements_by_class_name("artdeco-entity-lockup__subtitle")
        Company_followers.append(followers[0].text)
    except:
        Company_followers.append('NA')
```

```python
col=["designation","Company","location","posted_time","Applicant_count","Jobnature_level","Empcount_Ctype","Empcount","Emp_onLinkedin","Payrange","Companyfollowers"]
df=pd.DataFrame({"designation":Job_title,"Company":Company_name,"location":Location_name,"posted_time":Time_posted,"Applicant_count":Applicants_count,"Jobnature_level":Job_nature_level,"Empcount_Ctype":Emp_count_Ctype,"Empcount":Emp_count,"Emp_onLinkedin":Emp_on_Linkedin,"Payrange":Pay_range,"Companyfollowers":Company_followers})


```
